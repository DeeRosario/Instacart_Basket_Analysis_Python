# Grouping Data


```python
# import data
import os
import numpy as np
import pandas as pd
```

```python
# Set path
path = r'C:\Users\asus\Instacart Basket Analysis 01-2024'
```

```python
# import datafgrame from previous Task
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge_new1.pkl'))
```

```python
# create a subset
df = ords_prods_merge[:1000000]
```

```python
# checking the sape of the newly created dataframe

df.shape
```

```python
df.head(10)
```

```python
# group the dataframe by 'product_name'

df.groupby('product_name')
```

```python
# performing a single aggregation for the 'order_number' column

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# performing an aggregation using the mean command instead of the agg command

df.groupby('department_id')['order_number'].mean()
```

```python
# performing multiple aggregations simultaneously

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# creating a new column called 'max_order' in order to create a new flag

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# checking the values of the new column with head function

ords_prods_merge.head(100)
```

```python
# deriving new loyalty columns wiht loc function

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checking frequency of the new loyalty_flag column 

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# cheking multiple columns at the same time using head function

ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# turning outlier high values into NaNs since they don't make sense (from 4.9):

ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
